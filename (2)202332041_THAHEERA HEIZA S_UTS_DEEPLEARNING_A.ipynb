```python
def preprocess_image(image_path):
    """
    Fungsi untuk melakukan pra-pemrosesan citra tulisan tangan
    """
    try:
        # Baca gambar
        img = cv2.imread(image_path)

        if img is None:
            print(f"Warning: Tidak dapat membaca gambar {image_path}")
            return None, None

        # Konversi ke grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Inversi warna (background hitam, digit putih -> background putih, digit hitam)
        inverted = cv2.bitwise_not(gray)

        # Resize ke 28x28 piksel
        resized = cv2.resize(inverted, (28, 28), interpolation=cv2.INTER_AREA)

        # Normalisasi pixel values ke range [0, 1]
        normalized = resized / 255.0

        # Reshape untuk matching model input (28, 28, 1)
        reshaped = normalized.reshape(1, 28, 28, 1)

        return reshaped, normalized

    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")
        return None, None

def load_handwritten_dataset(base_dir):
    """
    Memuat dataset tulisan tangan dari folder
    """
    images = []
    labels = []
    image_paths = []
    original_images = []  # Untuk menyimpan gambar asli yang sudah diproses

    # Loop melalui setiap folder (0, 4, 1)
    for digit in ['0', '4', '1']:
        digit_dir = os.path.join(base_dir, digit)

        if os.path.exists(digit_dir):
            # Loop melalui setiap file dalam folder
            for filename in sorted(os.listdir(digit_dir)):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(digit_dir, filename)

                    # Pra-pemrosesan gambar
                    processed_img, original_processed = preprocess_image(image_path)

                    if processed_img is not None:
                        images.append(processed_img)
                        labels.append(int(digit))
                        image_paths.append(image_path)
                        original_images.append(original_processed)

    return images, labels, image_paths, original_images

# Memuat dataset tulisan tangan
print("Memuat dan memproses gambar tulisan tangan...")
handwritten_images, handwritten_labels, image_paths, original_processed_images = load_handwritten_dataset(base_dir)

print(f"Total gambar yang berhasil diproses: {len(handwritten_images)}")
print(f"Distribusi label: {np.unique(handwritten_labels, return_counts=True)}")
```

```python
# Visualisasi contoh citra sebelum dan setelah pra-pemrosesan
print("=== VISUALISASI CONTOH CITRA SETELAH PRA-PEMROSESAN ===")

# Tampilkan 6 contoh gambar
plt.figure(figsize=(15, 10))

for i in range(6):
    if i < len(handwritten_images):
        # Gambar asli (sebelum pra-pemrosesan)
        img_original = cv2.imread(image_paths[i])
        img_original_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)

        # Gambar setelah pra-pemrosesan
        img_processed = original_processed_images[i]

        # Plot gambar asli
        plt.subplot(3, 4, 2*i + 1)
        plt.imshow(img_original_rgb)
        plt.title(f'Asli: {handwritten_labels[i]}\n{os.path.basename(image_paths[i])}')
        plt.axis('off')

        # Plot gambar setelah pra-pemrosesan
        plt.subplot(3, 4, 2*i + 2)
        plt.imshow(img_processed, cmap='gray')
        plt.title(f'Setelah Pra-proses\n28x28, Grayscale')
        plt.axis('off')

plt.tight_layout()
plt.show()

# Visualisasi grid semua gambar setelah pra-pemrosesan
print("\n=== SEMUA GAMBAR SETELAH PRA-PEMROSESAN ===")
plt.figure(figsize=(20, 6))

for idx, (img_processed, label) in enumerate(zip(original_processed_images, handwritten_labels)):
    plt.subplot(3, 10, idx + 1)
    plt.imshow(img_processed, cmap='gray')
    plt.title(f'{label}', fontsize=10)
    plt.axis('off')

plt.suptitle('Semua Gambar Tulisan Tangan Setelah Pra-pemrosesan', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Cek jika model sudah ada, jika tidak buat model baru
try:
    print("Mencoba memuat model CNN yang sudah ditrain...")
    model = keras.models.load_model('mnist_cnn_model.h5')
    print("Model berhasil dimuat!")
except:
    print("Model tidak ditemukan. Melatih model CNN baru...")

    # Load dataset MNIST
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    # Preprocess data
    x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
    x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

    # Buat model CNN
    model = keras.Sequential([
        keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Conv2D(64, (3, 3), activation='relu'),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Conv2D(64, (3, 3), activation='relu'),
        keras.layers.Flatten(),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(10, activation='softmax')
    ])

    # Compile model
    model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

    # Train model
    history = model.fit(x_train, y_train,
                       epochs=5,
                       batch_size=128,
                       validation_split=0.1,
                       verbose=1)

    # Simpan model
    model.save('mnist_cnn_model.h5')
    print("Model baru berhasil dilatih dan disimpan!")

# Tampilkan summary model
model.summary()
```

```python
# Melakukan prediksi pada semua gambar tulisan tangan
print("Melakukan prediksi pada citra tulisan tangan...")
predictions = []
prediction_probs = []
prediction_confidences = []

for img in handwritten_images:
    pred_prob = model.predict(img, verbose=0)
    pred_class = np.argmax(pred_prob, axis=1)[0]
    confidence = np.max(pred_prob)

    predictions.append(pred_class)
    prediction_probs.append(pred_prob[0])
    prediction_confidences.append(confidence)

predictions = np.array(predictions)
true_labels = np.array(handwritten_labels)

# Menghitung akurasi
accuracy = accuracy_score(true_labels, predictions)
print(f"\n=== HASIL EVALUASI ===")
print(f"Akurasi Model pada Tulisan Tangan Sendiri: {accuracy:.4f} ({accuracy * 100:.2f}%)")
```

```python
# Menampilkan hasil prediksi untuk setiap citra dalam bentuk tabel
print("\n=== DETAIL HASIL PREDIKSI SETIAP CITRA ===")
print("\n" + "="*80)
print(f"{'No.':<3} {'File':<20} {'Label True':<10} {'Prediksi':<10} {'Confidence':<12} {'Status':<10}")
print("="*80)

correct_predictions = 0
total_predictions = len(handwritten_images)

for i, (true_label, pred_label, confidence, img_path) in enumerate(zip(true_labels, predictions, prediction_confidences, image_paths)):
    # Ekstrak nama file
    file_name = os.path.basename(img_path)
    folder_name = os.path.basename(os.path.dirname(img_path))

    status = "BENAR" if true_label == pred_label else "SALAH"
    if true_label == pred_label:
        correct_predictions += 1

    print(f"{i+1:<3} {file_name:<20} {true_label:<10} {pred_label:<10} {confidence:.4f}      {status:<10}")

print("="*80)
print(f"Total Prediksi Benar: {correct_predictions}/{total_predictions}")
print(f"Akurasi: {accuracy:.4f} ({accuracy * 100:.2f}%)")
```

```python
# Visualisasi hasil prediksi dengan warna berbeda untuk benar/salah
print("\n=== VISUALISASI HASIL PREDIKSI ===")

# Pisahkan prediksi benar dan salah
correct_indices = [i for i in range(len(predictions)) if predictions[i] == true_labels[i]]
wrong_indices = [i for i in range(len(predictions)) if predictions[i] != true_labels[i]]

print(f"Jumlah prediksi benar: {len(correct_indices)}")
print(f"Jumlah prediksi salah: {len(wrong_indices)}")

# Tampilkan semua gambar dengan marking benar/salah
plt.figure(figsize=(20, 12))

for idx, (img_processed, true_label, pred_label) in enumerate(zip(original_processed_images, true_labels, predictions)):
    plt.subplot(5, 6, idx + 1)

    # Gunakan warna hijau untuk benar, merah untuk salah
    cmap = 'viridis' if true_label == pred_label else 'hot'
    plt.imshow(img_processed, cmap=cmap)

    status = "✓" if true_label == pred_label else "✗"
    title_color = 'green' if true_label == pred_label else 'red'

    plt.title(f'True: {true_label} Pred: {pred_label} {status}',
              color=title_color, fontsize=10)
    plt.axis('off')

plt.suptitle('Hasil Prediksi Semua Gambar (Hijau: Benar, Merah: Salah)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Confusion Matrix
print("\n=== CONFUSION MATRIX ===")

cm = confusion_matrix(true_labels, predictions, labels=[0, 4, 1])

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='RdPu',
            xticklabels=['0', '4', '1'],
            yticklabels=['0', '4', '1'])
plt.title('Confusion Matrix - Evaluasi Tulisan Tangan Sendiri')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification report
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(true_labels, predictions, labels=[0, 4, 1], target_names=['0', '4', '1']))

# Analisis per digit
print("\n=== ANALISIS PER DIGIT ===")
for digit in [0, 4, 1]:
    digit_indices = [i for i, label in enumerate(true_labels) if label == digit]

    if digit_indices:
        digit_predictions = predictions[digit_indices]
        digit_accuracy = accuracy_score([digit] * len(digit_indices), digit_predictions)

        correct = sum(1 for pred in digit_predictions if pred == digit)

        print(f"\nDigit {digit}:")
        print(f"  - Jumlah sampel: {len(digit_indices)}")
        print(f"  - Akurasi: {digit_accuracy:.4f} ({digit_accuracy * 100:.2f}%)")
        print(f"  - Prediksi benar: {correct}/{len(digit_indices)}")

        # Tampilkan distribusi prediksi yang salah
        wrong_predictions = [pred for pred in digit_predictions if pred != digit]
        if wrong_predictions:
            unique_wrong, counts = np.unique(wrong_predictions, return_counts=True)
            print(f"  - Prediksi salah: {dict(zip(unique_wrong, counts))}")
```

```python
# Analisis confidence score
print("\n=== ANALISIS CONFIDENCE SCORE ===")

plt.figure(figsize=(12, 6))

# Plot confidence untuk prediksi benar dan salah
colors = ['green' if true == pred else 'red' for true, pred in zip(true_labels, predictions)]

plt.subplot(1, 2, 1)
plt.scatter(range(len(prediction_confidences)), prediction_confidences, c=colors, alpha=0.7)
plt.axhline(y=0.9, color='red', linestyle='--', alpha=0.7, label='Threshold 0.9')
plt.axhline(y=0.7, color='orange', linestyle='--', alpha=0.7, label='Threshold 0.7')
plt.xlabel('Sample Index')
plt.ylabel('Confidence Score')
plt.title('Confidence Score untuk Setiap Prediksi')
plt.legend()
plt.grid(True, alpha=0.3)

# Histogram confidence
plt.subplot(1, 2, 2)
correct_confidences = [conf for i, conf in enumerate(prediction_confidences) if true_labels[i] == predictions[i]]
wrong_confidences = [conf for i, conf in enumerate(prediction_confidences) if true_labels[i] != predictions[i]]

plt.hist(correct_confidences, bins=10, alpha=0.7, label='Benar', color='green')
plt.hist(wrong_confidences, bins=10, alpha=0.7, label='Salah', color='red')
plt.xlabel('Confidence Score')
plt.ylabel('Frequency')
plt.title('Distribusi Confidence Score')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Statistik confidence
print(f"Rata-rata confidence prediksi benar: {np.mean(correct_confidences):.4f}")
print(f"Rata-rata confidence prediksi salah: {np.mean(wrong_confidences):.4f}")
```
